### Question 1

Write a function in Python to check whether a number is Prime or not.

```python
def prime(num):
    if num > 1:
        for i in range(2,num):
            if(num%i == 0):
                print(num,'is not a prime number')
                break
        else:
            print(num,'is a prime a number')
    else:
        print(num,'is not a prime number')


num = int(input('Enter a number :'))
prime(num)
```
