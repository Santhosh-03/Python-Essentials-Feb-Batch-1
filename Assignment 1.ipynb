```python
my_list = [ 10, 20, 30, 40, 50, 60]
print('Original list is :',my_list)
```

#### Append()

```python
my_list = [ 10, 20, 30, 40, 50, 60]

print('Original list is :',my_list)
```

```python
my_list.append(70)

print('List after appending an element :',my_list)
```

#### Pop()

```python
my_list = [ 10, 20, 30, 40, 50, 60]

print('Original list is :',my_list)
```

```python
print('Poped element is :',my_list.pop())
```

#### Copy()

```python
my_list1 = [ 10, 20, 30, 40, 50, 60]

print('Original list is :',my_list)
```

```python
my_list2 = my_list1.copy()

print('Copied list is :',my_list2)
```

#### Insert ()

```python
my_list = [ 10, 20, 30, 40, 50, 60]

print('Original list is :',my_list)
```

```python
my_list.insert(2,25)

print('List after inserting an element :',my_list)
```

#### Remove()

```python
my_list = [ 10, 20, 30, 40, 50, 60]

print('Original list is :',my_list)
```

```python
my_list.remove(30)

print('After removing an element :',my_list)
```

#### Pop()

```python
car = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}

print('Dictionary is :',car)
```

```python
print('After poping the element :',car.pop('year'))
```

#### Update()

```python
car = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}

print('Dictionary is :',car)
```

```python
car.update({'color':'White'})

print('After updating the dictionary :\n',car)
```
